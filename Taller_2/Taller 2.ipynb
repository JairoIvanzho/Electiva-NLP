# Punto 2:

- Hacer Web Scraping de 10 biografías en Wikipedia (en búcle)
- Obtener el **encabezado** de cada biografía
- Obtener todos los **contenidos** y **etiquetas de título** asociados a los **links** del primer párrafo

<img src='img/carlsagan.png'>

```python
#Asignamos las 10 biografías que queremos encontrar:
urls = ['https://es.wikipedia.org/wiki/Carlos_Vives'
,'https://es.wikipedia.org/wiki/Juanes'
,'https://es.wikipedia.org/wiki/Andr%C3%A9s_Cepeda_(cantante)'
,'https://es.wikipedia.org/wiki/Sim%C3%B3n_Bol%C3%ADvar'
,'https://es.wikipedia.org/wiki/Policarpa_Salavarrieta'
,'https://es.wikipedia.org/wiki/Mariana_Paj%C3%B3n'
,'https://es.wikipedia.org/wiki/Carlos_Valderrama'
,'https://es.wikipedia.org/wiki/Faustino_Asprilla'
,'https://es.wikipedia.org/wiki/Ren%C3%A9_Higuita'
,'https://es.wikipedia.org/wiki/Radamel_Falcao_Garc%C3%ADa']
```

```python
#Obtenemos los encabezados de todos y cada una de las biografías
for linkBusqueda in urls:
    request = urllib.request.urlopen(linkBusqueda)
    fuente = request.read()
    request.close()
    soup = bs.BeautifulSoup(fuente, 'html.parser')
    print(soup.find(id='firstHeading').contents)
```

```python
for linkBusqueda in urls:
    
    request = urllib.request.urlopen(linkBusqueda)
    fuente = request.read()
    request.close()
    soup = bs.BeautifulSoup(fuente, 'html.parser')
    print('Encabezado:')
    print(soup.find(id='firstHeading').contents)
    print()
    for linkParrafo in soup.find(id="mw-content-text").find('p').find_all('a'):
        #print (linkParrafo)
        print ('Contenido:')
        print (linkParrafo.contents)
        print ('Título:')
        print(linkParrafo.get('title'))
        print()
    print("*"*100)
    print()
```

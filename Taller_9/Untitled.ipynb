```python
import numpy as np

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('english')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
with open("./CEO_Messages.txt", encoding="utf-8") as archivo:
    data = archivo.read()
print ("En el archivo hay: " + str(len(data.split())) + " Palabras")
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='white',
                     colormap='gist_heat',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
eriLogo = PIL.Image.open("Ericsson_Logo.jpg")
display(eriLogo)
```

```python
eriMask = np.array(eriLogo)
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='white',
                     colormap='copper_r',
                      mask=eriMask, contour_width=0.5, contour_color='black',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

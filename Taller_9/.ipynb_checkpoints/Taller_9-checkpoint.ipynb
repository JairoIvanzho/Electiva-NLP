```python
import numpy as np

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('english')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

Cargamos los datos de un archivo de texto donde se encuentran los últimos 3 mensajes enviados por el CEO de Ericsson a todos los colaboradores.

```python
with open("./CEO_Messages.txt", encoding="utf-8") as archivo:
    data = archivo.read()
print ("En el archivo hay: " + str(len(data.split())) + " Palabras")
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='white',
                     colormap='gist_heat',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

Nuestra nube de palabras nos muestra que, obviamente la palabra Ericsson es la que más utiliza, sin embargo, también utiliza palabras como "important", "vision", "team", "decision", "need", etc.

Ahora, cargamos el logo de Ericsson para hacer la nube de palabras en el logo:

```python
eriLogo = PIL.Image.open("ericssonLogo.png")
display(eriLogo)
```

```python
eriMask = np.array(eriLogo)
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='white',
                     colormap='seismic',
                      mask=eriMask, contour_width=0.5, contour_color='black',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
